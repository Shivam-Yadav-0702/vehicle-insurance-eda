
# Exploratory Data Analysis (EDA) on Vehicle Insurance Dataset

## Project Objective
The main goal of this project is to perform Exploratory Data Analysis (EDA) on a Vehicle Insurance dataset to understand customer behavior, insurance claim patterns, and factors influencing policy responses.

This notebook covers:
- Data Cleaning
- Missing Value Analysis
- Univariate & Bivariate Analysis
- Premium Distribution
- Age & Gender Analysis
- Vehicle Damage Impact
- Region-wise Insights
- Claim Frequency Analysis
- Business Insights & Recommendations

---

## About Dataset
The dataset contains information about customers such as:
- Age
- Gender
- Vehicle Age
- Vehicle Damage History
- Insurance Premium
- Region
- Driving License
- Previously Insured Status
- Response (Target Variable)

The target column:
- **Response = 1** → Customer interested in vehicle insurance
- **Response = 0** → Customer not interested

---

### Author Notes
This project was built as a practical EDA case study using Python libraries like Pandas, Matplotlib, and Seaborn.


```

# Import required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

# Display settings
pd.set_option('display.max_columns', None)

# Load dataset
df = pd.read_csv('Vehicle_Insurance.csv')

print("Dataset Loaded Successfully")
```

```

# Basic dataset overview

print("Shape of dataset:", df.shape)

print("\nColumns in dataset:\n")
print(df.columns)

print("\nFirst 5 Rows:\n")
display(df.head())

print("\nDataset Information:\n")
df.info()
```

```

# Missing value analysis

missing_values = df.isnull().sum()

print("Missing Values in Dataset:\n")
print(missing_values)

# Visualizing missing values
plt.figure(figsize=(10,4))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False)
plt.title("Missing Value Heatmap")
plt.show()
```

```

# Statistical summary

display(df.describe())

print("\nCategorical Columns Summary:\n")
display(df.describe(include='object'))
```


# Understanding Target Variable

The target variable `Response` shows whether a customer is interested in vehicle insurance.

- 0 → Not Interested
- 1 → Interested


```

# Target variable distribution

plt.figure(figsize=(6,4))

sns.countplot(x='Response', data=df)

plt.title("Response Distribution")
plt.xlabel("Response")
plt.ylabel("Count")

plt.show()

response_percent = df['Response'].value_counts(normalize=True)*100
print(response_percent)
```

```

# Age distribution analysis

plt.figure(figsize=(10,5))

sns.histplot(df['Age'], bins=30, kde=True)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()

print("Average Age:", round(df['Age'].mean(),2))
```

```

# Gender analysis

plt.figure(figsize=(6,4))

sns.countplot(x='Gender', data=df)

plt.title("Gender Distribution")
plt.show()

# Response by Gender
plt.figure(figsize=(6,4))

sns.countplot(x='Gender', hue='Response', data=df)

plt.title("Response by Gender")
plt.show()
```

```

# Annual Premium Distribution

plt.figure(figsize=(10,5))

sns.boxplot(x=df['Annual_Premium'])

plt.title("Annual Premium Distribution")
plt.show()

plt.figure(figsize=(10,5))

sns.histplot(df['Annual_Premium'], bins=50)

plt.title("Annual Premium Histogram")
plt.show()
```

```

# Vehicle Age Analysis

plt.figure(figsize=(7,4))

sns.countplot(x='Vehicle_Age', data=df)

plt.title("Vehicle Age Distribution")
plt.show()

# Vehicle Age vs Response
plt.figure(figsize=(7,4))

sns.countplot(x='Vehicle_Age', hue='Response', data=df)

plt.title("Vehicle Age vs Response")
plt.show()
```

```

# Vehicle Damage Analysis

plt.figure(figsize=(6,4))

sns.countplot(x='Vehicle_Damage', hue='Response', data=df)

plt.title("Vehicle Damage vs Response")
plt.show()

damage_response = pd.crosstab(df['Vehicle_Damage'], df['Response'], normalize='index')*100

print(damage_response)
```

```

# Previously insured customers

plt.figure(figsize=(6,4))

sns.countplot(x='Previously_Insured', hue='Response', data=df)

plt.title("Previously Insured vs Response")
plt.show()
```

```

# Top regions analysis

top_regions = df['Region_Code'].value_counts().head(10)

plt.figure(figsize=(12,5))

sns.barplot(x=top_regions.index, y=top_regions.values)

plt.title("Top 10 Regions")
plt.xlabel("Region Code")
plt.ylabel("Count")

plt.show()
```

```

# Correlation heatmap

numeric_df = df.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(12,8))

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```


# Key Insights from Analysis

## Important Findings

### 1. Response Imbalance
Most customers are not interested in vehicle insurance, which shows a clear class imbalance.

### 2. Age Impact
Middle-aged customers show higher interest in purchasing insurance compared to younger users.

### 3. Vehicle Damage
Customers with previous vehicle damage are much more likely to buy insurance.

### 4. Previously Insured Customers
Customers already insured rarely respond positively to new insurance offers.

### 5. Vehicle Age
Older vehicles tend to have higher insurance interest rates.

### 6. Premium Distribution
Annual premiums contain outliers, which is common in insurance datasets.

---

# Business Recommendations

- Focus marketing on customers with damaged vehicles.
- Target middle-aged users for better conversion rates.
- Create special offers for owners of older vehicles.
- Use segmentation strategies based on insurance history.
